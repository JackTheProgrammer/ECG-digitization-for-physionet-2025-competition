```python
from pathlib import Path

base_dir = Path('D:/ml projects/ecg_segmentation')
output_dir = base_dir / 'physionet_ecg_graphs_images'
if output_dir.exists():
    print(f"Directory {output_dir} already exists.")
else:
    output_dir.mkdir(parents=True, exist_ok=False)
```

```python
import os, shutil

train_imgs_dir = '../physionet_ecg_competition_2025/train'
output_dir = '../physionet_ecg_graphs_images/'
image_folders = []

for image_folder in os.listdir(train_imgs_dir):
    image_folders.append(image_folder)
    print(" Processing folder:", image_folder)
    image_folder_path = f"{train_imgs_dir}/{image_folder}"
    if not os.path.isdir(image_folder_path):
        continue
    print(" Created directory:", f"{output_dir}/{image_folder}")
    for image_file in os.listdir(image_folder_path):
        if image_file.endswith('.png') and image_file.split('.')[0].split('-')[1] == '0001':
            print("  Processing file:", image_file)
            src_path = f"{image_folder_path}/{image_file}"
            dst_path = f"{output_dir}/{image_file}"
            try:
                shutil.copy2(src_path, dst_path)
                print(f"  Copied {src_path} to {dst_path}")
            except IOError as e:
                print(f"  Unable to copy file. {e}")
            except Exception as e:
                print(f"  Unexpected error: {e}")

print(f"Dealt with {len(image_folders)} folders.")
```
